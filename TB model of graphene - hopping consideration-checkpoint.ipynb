```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
import math as m
import cmath
from matplotlib.path import Path
from matplotlib.patches import PathPatch
a = 0.5 # basis in real space
```

```python
def hopping(length):
    return (2.934478)*m.exp(-7.5*((m.sqrt(3)*length/a) - 1.0))
print(hopping(m.sqrt(3)*a))
```

```python
E_onsite = -0.880586
Etunnel1 = np.zeros((3), float)
Etunnel1[:] = (-1)*hopping(a/m.sqrt(3))
Etunnel2 = np.zeros((6), float)
Etunnel2[:] = hopping(m.sqrt(3)*a)
Etunnel3 = np.zeros((3), float)
Etunnel3[:] = (-1)*hopping(2.0*a)
#d = a_real**2/(2*t1[0])    # distance between two atoms
print(Etunnel1, '\n', Etunnel2, '\n', Etunnel3)
```

```python
# 形變
def Delta(t1, t2, delta, ratio):
    ratio = 1.0/(1+ratio)
    d = (LA.norm(t1))**2/(2*t1[0])
    delta = delta*d
    dvec = np.array([(d + delta), 0.], float)
    p1 = t1 - np.array([d, 0])
    p2 = t2 - np.array([d, 0])
    p1 *= (LA.norm(p1) - delta*ratio)/(LA.norm(p1))
    p2 *= (LA.norm(p2) - delta*(1.0-ratio))/(LA.norm(p2))
    t1 = dvec + p1
    t2 = dvec + p2
    ls = [t1, t2]
    ls = np.array(ls, float)
    return ls

def DisformHopping(NN, Eh):
    # Total length is invariant
    E = [0.0 for i in range(len(Eh))]
    if len(E) == 3:
        for i in range(len(E)):
            E[i] = (-1)*hopping(LA.norm(NN[i]))
    else:
        for i in range(len(E)):
            E[i] = hopping(LA.norm(NN[i]))
    return E
```

```python
trans1 = np.array([m.sqrt(3)*a/2.0, a/2.0])
trans2 = trans1.copy()
trans2[1] = (-1)*trans2[1]
# Distance between atoms
d = (LA.norm(trans1))**2/(2*trans1[0])
print(d)
print(trans1, trans2)
delta = -0.07
ratio = 1.0
ls = Delta(trans1, trans2, delta, ratio)
t1 = ls[0]
t2 = ls[1]
print("Translation basis in real: ", t1, ", ", t2)
print(trans1, trans2)

def NearNeibor(t1, t2):
    dvec = np.array([(t1[0] - t1[1]/m.sqrt(3)), 0], float)
    secondNN = [t1, (t1-t2), -t2, -t1, (t2-t1), t2]
    firstNN = [dvec, (-t2 + dvec), (-t1 + dvec)]
    thirdNN = [(t1-t2)+dvec, (-t2-t1)+dvec, (-t1+t2)+dvec]
    return [firstNN, secondNN, thirdNN]
    
#print(nearestA)
#print(nearestB)
# 2D case
def reciprocalGen(t1, t2):
    g1 = np.array([t2[1], (-1)*t2[0]])
    g1 = g1*2.0*(m.pi)/(g1.dot(t1))
    g2 = np.array([t1[1], (-1)*t1[0]])
    g2 = g2*2.0*(m.pi)/(g2.dot(t2))
    return [g1, g2]
[reci1, reci2] = reciprocalGen(t1, t2)
print("reciprocal: ", reci1, ', ', reci2)

print(Etunnel1)
hop1 = DisformHopping(NearNeibor(t1, t2)[0], Etunnel1)
print(hop1)
print("Check:", abs(hop1[0]), abs(hop1[1])+abs(hop1[2]))
dirac = True
if (abs(hop1[1])+abs(hop1[2])) < abs(hop1[0]):
    dirac = False
print(hop1[0]/hop1[1])
print(hop1[2]/hop1[1])
```

```python
def solveK(d1, d2, P1, P2):
    KY = (d2[0]*P1 - d1[0]*P2)/(d2[0]*d1[1] - d1[0]*d2[1])
    KX = (d2[1]*P1 - d1[1]*P2)/(d2[1]*d1[0] - d1[1]*d2[0])
    return np.array([KX, KY])
def minVal(deform, ratio):
    [t1, t2] = Delta(trans1, trans2, deform, ratio)
    NN1 = NearNeibor(t1, t2)[0]
    h1 = DisformHopping(NN1, Etunnel1)
    [g1, g2] = reciprocalGen(t1, t2)

    if (abs(h1[1])+abs(h1[2])) >= abs(h1[0]):
        alpha1 = (h1[2]**2.0 + h1[0]**2.0 - h1[1]**2.0)
        alpha1 /= 2.0*(h1[0]*h1[2])
        #print("alpha1 =", alpha1)
        alpha1 = m.acos(alpha1)
    else:
        alpha1 = 2*m.pi    # 2n(pi)       
    K1 = alpha1 - m.pi
    #print("K1:", K1) 

    if  (abs(h1[1])+abs(h1[2])) >= abs(h1[0]):
        alpha2 = (h1[0]**2.0 + h1[1]**2.0 - h1[2]**2.0)
        alpha2 /= 2.0*h1[0]*h1[1]
        #print("alpha2 =", alpha2)
        alpha2 = m.acos(alpha2)
    else:
        alpha2 = 0.0*m.pi      
    K2 = -m.pi - alpha2
    #print("K2:", K2)
    #print("Check1:", Etunnel1[0], Etunnel1[1]*m.cos(alpha1)+Etunnel1[2]*m.cos(alpha2))
    #print("Check2:", Etunnel1[1]*m.sin(alpha1)-Etunnel1[2]*m.sin(alpha2))

    kmin = [0 for i in range(6)]
    # inverse is also a valid solution
    kmin[0] = solveK(t1, t2, K1, K2)
    kmin[3] = solveK(t1, t2, -K1, -K2)
    kmin[4] = solveK(t1, t2, K1+2.0*m.pi, K2+2.0*m.pi)
    kmin[1] = kmin[3] - (g1+g2)
    kmin[2] = kmin[0] + g2
    kmin[5] = kmin[3] - g2
    kmin = np.array(kmin)
    '''
    for i in range(6):
        temp = 0.0
        print(kmin[i])
        for n in range(len(NN1)):
            temp += h1[n]*cmath.exp(1j*kmin[i].dot(NN1[n]))
        print("check", i+1, ":", temp)
    '''
    return kmin
#print(minVal(0.41, 2.0))
k_min = minVal(delta, ratio)
print(k_min)
```

```python
# Hamiltonian consider nearest 3 atoms
def H_nearB(path, i, Et1, NN1):
    H = np.empty((2, 2), complex)
    H[0][0] = E_onsite
    H[1][1] = H[0][0]
    # Calculate H_01
    temp = 0.0
    for n in range(len(NN1)):
        temp += Et1[n]*cmath.exp(1j*path[i].dot(NN1[n]))
    H[0][1] = temp
    H[1][0] = H[0][1].real - 1j*H[0][1].imag
    #print(H)
    #print(LA.eigvals(H))
    return H

# Hamiltonian consider nearest 3+6 atoms
def H_nearAB(path, i, Et1, Et2, NN1, NN2):
    H = np.empty((2, 2), complex)
    # Calculate diagonal term
    temp = 0.0
    for n in range(len(NN2)):
        temp += Et2[n]*cmath.exp(1j*path[i].dot(NN2[n]))
    H[0][0] = E_onsite + temp
    H[1][1] = H[0][0]
    # Calculate off-iagonal term
    temp = 0.0
    for n in range(len(NN1)):
        temp += Et1[n]*cmath.exp(1j*path[i].dot(NN1[n]))
    H[0][1] = temp
    H[1][0] = H[0][1].real - 1j*H[0][1].imag
    #print(H)
    #print(LA.eigvals(H))
    return H

# Hamiltonian consider nearest 3+6+3 atoms
def H_nearBAB(path, i, Et1, Et2, Et3, NN1, NN2, NN3):
    H = np.empty((2, 2), complex)
    # Calculate diagonal term
    temp = 0.0
    for n in range(len(NN2)):
        temp += Et2[n]*cmath.exp(1j*path[i].dot(NN2[n]))
    H[0][0] = E_onsite + temp
    H[1][1] = H[0][0]
    # Calculate H_01
    paraB1 = 0.0
    paraB2 = 0.0
    for n in range(len(NN1)):
        paraB1 += Et1[n]*cmath.exp(1j*path[i].dot(NN1[n]))
    for n in range(len(NN3)):
        paraB2 += Et3[n]*cmath.exp(1j*path[i].dot(NN3[n]))
    H[0][1] = paraB1 + paraB2
    H[1][0] = H[0][1].real - 1j*H[0][1].imag
    #print(H)
    #print(LA.eigvals(H))
    return H
```

```python
def eigval_B(k, Et1, NN1):
    # Eigenvalue
    eig1_r = np.zeros(k.shape[0])
    eig2_r = np.zeros(k.shape[0])
    '''
    # Eigenvector
    c2A = np.empty((k.shape[0], 1), float)
    c2B = np.empty((k.shape[0], 1), float)
    '''
    for i in range(k.shape[0]):
        H = H_nearB(k, i, Et1, NN1)
        #E, v = LA.eig(H_nearB(k, i))
        eig1_r[i] = H[0][0].real + m.sqrt(H[0][1].real**2 + H[0][1].imag**2)
        eig2_r[i] = H[0][0].real - m.sqrt(H[0][1].real**2 + H[0][1].imag**2)
    return [eig1_r, eig2_r]

def eigval_AB(k, Et1, Et2, NN1, NN2):
    eig1_r = np.zeros(k.shape[0])
    eig2_r = np.zeros(k.shape[0])
    '''
    cA = np.empty((k.shape[0], 1), complex)
    cB = np.empty((k.shape[0], 1), complex)
    '''
    for i in range(k.shape[0]):
        H = H_nearAB(k, i, Et1, Et2, NN1, NN2)
        #E, v = LA.eig(H_nearAB(k, i))
        eig1_r[i] = H[0][0].real + m.sqrt(H[0][1].real**2 + H[0][1].imag**2)
        eig2_r[i] = H[0][0].real - m.sqrt(H[0][1].real**2 + H[0][1].imag**2)
    return [eig1_r, eig2_r]

def eigval_BAB(k, Et1, Et2, Et3, NN1, NN2, NN3):
    eig1_r = np.zeros(k.shape[0])
    eig2_r = np.zeros(k.shape[0])
    for i in range(k.shape[0]):
        H = H_nearBAB(k, i, Et1, Et2, Et3, NN1, NN2, NN3)
        #E, v = LA.eig(H_nearBAB(k, i))
        eig1_r[i] = H[0][0].real + m.sqrt(H[0][1].real**2 + H[0][1].imag**2)
        eig2_r[i] = H[0][0].real - m.sqrt(H[0][1].real**2 + H[0][1].imag**2)
    return [eig1_r, eig2_r]
```
