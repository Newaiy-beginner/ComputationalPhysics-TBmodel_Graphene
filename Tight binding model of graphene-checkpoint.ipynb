```python
# 形變
def Delta(t1, t2, delta, ratio):
    ratio = 1.0/(1+ratio)
    d = (LA.norm(t1))**2/(2*t1[0])
    delta = delta*d
    dvec = np.array([(d + delta), 0.], float)
    p1 = t1 - np.array([d, 0])
    p2 = t2 - np.array([d, 0])
    p1 *= (LA.norm(p1) - delta*ratio)/(LA.norm(p1))
    p2 *= (LA.norm(p2) - delta*(1.0-ratio))/(LA.norm(p2))
    t1 = dvec + p1
    t2 = dvec + p2
    ls = [t1, t2]
    ls = np.array(ls, float)
    return ls
```

```python
def DisformHopping(NN, Eh):
    # Total length is invariant
    E = [0.0 for i in range(len(Eh))]
    Sum = 0.
    for i in range(len(NN)):
        Sum += LA.norm(NN[i])
    for i in range(len(NN)):
        delta = (LA.norm(NN[i]) - Sum/len(NN))/(Sum/len(NN))
        E[i] = Eh[i] + delta*abs(Eh[i])
    return E
print(Etunnel1)
print(Etunnel2)
print(Etunnel3)
print(DisformHopping(NearNeibor(t1, t2)[0], Etunnel1))
print(DisformHopping(NearNeibor(t1, t2)[1], Etunnel2))
print(DisformHopping(NearNeibor(t1, t2)[2], Etunnel3))
hop1 = DisformHopping(NearNeibor(t1, t2)[0], Etunnel1)
print(hop1[0]/hop1[1])
print(hop1[2]/hop1[1])
```

```python
# Hamiltonian consider nearest 3 atoms
def H_nearB(path, i, Et1, NN1):
    H = np.empty((2, 2), complex)
    H[0][0] = E_onsite
    H[1][1] = H[0][0]
    # Calculate H_01
    temp = 0.0
    for n in range(len(NN1)):
        temp += Et1[n]*cmath.exp(1j*path[i].dot(NN1[n]))
    H[0][1] = temp
    H[1][0] = H[0][1].real - 1j*H[0][1].imag
    #print(H)
    #print(LA.eigvals(H))
    return H

# Hamiltonian consider nearest 3+6 atoms
def H_nearAB(path, i, Et1, Et2, NN1, NN2):
    H = np.empty((2, 2), complex)
    # Calculate diagonal term
    temp = 0.0
    for n in range(len(NN2)):
        temp += Et2[n]*cmath.exp(1j*path[i].dot(NN2[n]))
    H[0][0] = E_onsite + temp
    H[1][1] = H[0][0]
    # Calculate off-iagonal term
    temp = 0.0
    for n in range(len(NN1)):
        temp += Et1[n]*cmath.exp(1j*path[i].dot(NN1[n]))
    H[0][1] = temp
    H[1][0] = H[0][1].real - 1j*H[0][1].imag
    #print(H)
    #print(LA.eigvals(H))
    return H

# Hamiltonian consider nearest 3+6+3 atoms
def H_nearBAB(path, i, Et1, Et2, Et3, NN1, NN2, NN3):
    H = np.empty((2, 2), complex)
    # Calculate diagonal term
    temp = 0.0
    for n in range(len(NN2)):
        temp += Et2[n]*cmath.exp(1j*path[i].dot(NN2[n]))
    H[0][0] = E_onsite + temp
    H[1][1] = H[0][0]
    # Calculate H_01
    paraB1 = 0.0
    paraB2 = 0.0
    for n in range(len(NN1)):
        paraB1 += Et1[n]*cmath.exp(1j*path[i].dot(NN1[n]))
    for n in range(len(NN3)):
        paraB2 += Et3[n]*cmath.exp(1j*path[i].dot(NN3[n]))
    H[0][1] = paraB1 + paraB2
    H[1][0] = H[0][1].real - 1j*H[0][1].imag
    #print(H)
    #print(LA.eigvals(H))
    return H
```

```python
def eigval_B(k, Et1, NN1):
    # Eigenvalue
    eig1_r = np.zeros(k.shape[0])
    eig2_r = np.zeros(k.shape[0])
    '''
    # Eigenvector
    c2A = np.empty((k.shape[0], 1), float)
    c2B = np.empty((k.shape[0], 1), float)
    '''
    for i in range(k.shape[0]):
        H = H_nearB(k, i, Et1, NN1)
        #E, v = LA.eig(H_nearB(k, i))
        eig1_r[i] = H[0][0].real + m.sqrt(H[0][1].real**2 + H[0][1].imag**2)
        eig2_r[i] = H[0][0].real - m.sqrt(H[0][1].real**2 + H[0][1].imag**2)
    return [eig1_r, eig2_r]

def eigval_AB(k, Et1, Et2, NN1, NN2):
    eig1_r = np.zeros(k.shape[0])
    eig2_r = np.zeros(k.shape[0])
    '''
    cA = np.empty((k.shape[0], 1), complex)
    cB = np.empty((k.shape[0], 1), complex)
    '''
    for i in range(k.shape[0]):
        H = H_nearAB(k, i, Et1, Et2, NN1, NN2)
        #E, v = LA.eig(H_nearAB(k, i))
        eig1_r[i] = H[0][0].real + m.sqrt(H[0][1].real**2 + H[0][1].imag**2)
        eig2_r[i] = H[0][0].real - m.sqrt(H[0][1].real**2 + H[0][1].imag**2)
    return [eig1_r, eig2_r]

def eigval_BAB(k, Et1, Et2, Et3, NN1, NN2, NN3):
    eig1_r = np.zeros(k.shape[0])
    eig2_r = np.zeros(k.shape[0])
    for i in range(k.shape[0]):
        H = H_nearBAB(k, i, Et1, Et2, Et3, NN1, NN2, NN3)
        #E, v = LA.eig(H_nearBAB(k, i))
        eig1_r[i] = H[0][0].real + m.sqrt(H[0][1].real**2 + H[0][1].imag**2)
        eig2_r[i] = H[0][0].real - m.sqrt(H[0][1].real**2 + H[0][1].imag**2)
    return [eig1_r, eig2_r]
```
